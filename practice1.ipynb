```python
import cv2
import matplotlib.pyplot as plt

# Assuming 'img1.jpg' is in the current directory
image1 = cv2.imread('image2.jpg', 1)

# Convert BGR (OpenCV default) to RGB (Matplotlib default)
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

plt.imshow(image1)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Assuming 'img1.jpg' is in the current directory
image1 = cv2.imread('image2.jpg', 1)
color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.calcHist([image1], [i], None, [256], [0, 256])
    plt.plot(histr, color=col)

plt.xlim([0, 256])
plt.title('Combined RGB Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```

```python
height, width, _ = image1.shape # unpack the three values and discard the third (channels)
for i in range(0,height-1):
  for j in range(0,width-1):
    pixel=image1[i,j]
    pixel[0]=255-pixel[0]
    pixel[1]=255-pixel[1]
    pixel[2]=255-pixel[2]
    image1[i,j]=pixel

plt.imshow(image1)
plt.show()
```

```python
color = ('b', 'g', 'r')

for i, col in enumerate(color):

    histr = cv2.calcHist([image1],
                         [i], None,
                         [256],
                         [0, 256])

    plt.plot(histr, color = col)
    plt.xlim([0, 256])

plt.show()
```
